```python
df_erobis = pd.read_csv("ER_OBIS.csv",low_memory=False) # data downloaded from OBIS webside, see more info in README 
lat_erobis = df_erobis.decimalLatitude
lon_erobis = df_erobis.decimalLongitude

## deal with different formats of eventDate
df_erobis['eventDate'] = pd.to_datetime(df_erobis['eventDate'], format='mixed', dayfirst=True, errors='coerce')
df_erobis = df_erobis[['decimalLatitude', 'decimalLongitude','eventDate','occurrenceStatus','organismQuantity','organismQuantityType','date_year','sst','sss']]
```

```python
group_df = df_erobis.groupby('date_year', group_keys=True)
```

```python
## Rounding to 0.25 degrees
def round_coordinates(coord):
    return np.round(coord * 4) / 4  
```

```python
# Apply the rounding function to latitude and longitude columns
rounded_df = group_df.apply(lambda x: x.assign(Latitude=x['decimalLatitude'].apply(round_coordinates),
                                                    Longitude=x['decimalLongitude'].apply(round_coordinates)))
# Reset index on the resulting DataFrame
rounded_df.reset_index(inplace=True, drop=True)
```

```python
# Group the data by rounded latitude and longitude and count occurrences
df_summerised = rounded_df.groupby(['date_year', 'Latitude', 'Longitude']).size().reset_index(name='Occurrences')  #
df_summerised = df_summerised[(df_summerised.date_year > 1980)&(df_summerised.date_year < 2023) & (df_summerised.Longitude > 110) & (df_summerised.Longitude < 160)]
# Reset the index of df_summerised starting from 0
df_summerised.reset_index(drop=True, inplace=True)
df_summerised['date_year'] = df_summerised['date_year'].astype(int)
```

```python
df_summerised = df_summerised.groupby(['Latitude', 'Longitude']).size().reset_index(name='counts')
df_summerised
```
